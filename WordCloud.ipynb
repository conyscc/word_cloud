```python
with open('Уэллс_-_Война_миров.txt', 'r') as file:
    data = file.readlines()
```

```python
import numpy as np
from nltk.tokenize import RegexpTokenizer
```

```python
tokenizer = RegexpTokenizer(r'\w+')
data = np.array(data)
words = tokenizer.tokenize(''.join(data).lower())
```

```python
print('Кол-во строк в файле =', data.size)
print('Кол-во слов =', len(words))
print('Кол-во символов без пробелов =', len("".join(words)))
print('Кол-во символов включая пробелы =', len(" ".join(words)))
```

```python
from nltk.corpus import stopwords
import pandas as pd
```

```python
stopwrds = set(stopwords.words('russian'))
filtered_words = [w for w in words if w not in stopwrds]
words_series = pd.Series(filtered_words)
words_dict = dict(words_series.value_counts()[:80])
```

```python
from wordcloud import WordCloud
from PIL import Image
import matplotlib.pyplot as plt
```

```python
mask = np.array(Image.open('sleep.jpg'))
wordcloud = WordCloud(colormap ='Pastel1', mask = mask)
wordcloud.generate_from_frequencies(frequencies = words_dict)
plt.figure(figsize = (10, 7))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
